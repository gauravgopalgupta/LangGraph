```python
from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
from langchain_core.messages import HumanMessage
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python

# 1. Setup Token
#os.environ["HUGGINGFACEHUB_API_TOKEN"] = "replace with your actual token"
#load_dotenv()  # load environment variables from .env file
# 2. Define the Endpoint
# Note: We keep task as "text-generation" here because that's what the 
# HF library expects, but we wrap it in ChatHuggingFace to satisfy the provider.
# 2. Use a currently supported model on HF Serverless
# Mistral-7B-Instruct-v0.2 is very stable on HF endpoints
llm = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2", 
    task="text-generation",
    max_new_tokens=512,
)

# 3. Wrap in ChatHuggingFace
chat_model = ChatHuggingFace(llm=llm)

# 4. Invoke
response = chat_model.invoke([HumanMessage(content="Who is the prime minister of India?")])
print(response.content)
```
